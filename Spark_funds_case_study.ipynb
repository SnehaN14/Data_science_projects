```python
#As we can see the in the master frame there are different sectors containing value for particular primary sector
#we need to do conversion from wide to long format using pd.melt

value_vars=master_frame.columns[9:18]

id_vars=master_frame.columns.difference(value_vars)
value_vars
```

```python
id_vars
```

```python
final_df = pd.melt(master_frame, 
        id_vars=list(id_vars), 
        value_vars=list(value_vars))

final_df.head()
```

```python
final_df = final_df[final_df['value']==1]
final_df = final_df.drop(columns='value')
final_df.head()
```

```python
final_df=final_df.rename(columns={'variable':'main sector'})
final_df.head()
```

```python
final_df=final_df[(final_df.raised_amount_usd >=5000000) & (final_df.raised_amount_usd <=15000000)]
final_df.head()
```

```python
final_df.isnull().sum()
```

```python
final_df=final_df.rename(columns={'main sector':'main_sector'})
df_2=final_df.groupby(['country_code', 'main_sector']).raised_amount_usd.agg(['count', 'sum'])  
df_2
```
