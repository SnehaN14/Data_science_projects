```python
df_categorical2 = df.select_dtypes(include=['int64','float64'])
df_categorical2.head()
```

```python
df_categorical2.drop('id', axis=1, inplace=True)
```

```python
df_categorical2.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
nc_columns=list(df_categorical2.columns)
```

```python
df_categorical2[nc_columns] = scaler.fit_transform(df_categorical2[nc_columns])
```

```python
df_categorical2.head()
```

```python
df.drop('id',axis=1, inplace=True)
```

```python
df.drop(nc_columns, axis=1, inplace=True)
```

```python
df.head()
```

```python
final_df= pd.concat([df,df_categorical2], axis=1)
```

```python
final_df.head()
```

```python
final_df['diagnosis'] = df['diagnosis'].apply({'B':0, 'M':1}.get)
final_df['diagnosis'].head()
```

```python
# Importing test_train_split from sklearn library
from sklearn.model_selection import train_test_split
y=final_df["diagnosis"]
X=final_df.drop(['diagnosis'],axis=1)
```

```python
#Splitting data into train and test data
X_train,X_test,y_train,y_test = train_test_split(X,y,train_size = 0.7, test_size= 0.3, random_state = 100)
```

```python
#Running Your First Training Model
import statsmodels.api as sm
```

```python
# Logistic regression model
logm1 = sm.GLM(y_train,(sm.add_constant(X_train)), family = sm.families.Binomial())
#logm1.fit().summary()
```

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
from sklearn.feature_selection import RFE
rfe = RFE(logreg, 10)             # running RFE with 15 variables as output
rfe = rfe.fit(X_train,y_train)   
```

```python
list(zip(X_train.columns, rfe.support_, rfe.ranking_))
```

```python
col = X_train.columns[rfe.support_]
```

```python
X_train_sm=sm.add_constant(X_train[col])
logm2 = sm.GLM(y_train,X_train_sm, family = sm.families.Binomial())
modres = logm2.fit()
logm2.fit().summary()
```

```python
#predicted  values
# Getting the predicted values on the train set
y_train_pred = modres.predict(X_train_sm)
y_train_pred[:10]
```

```python
import pandas as pd
import numpy as np
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import scale
```

```python
# using rbf kernel, C=1, default value of gamma
#try  SVM
model = SVC(C = 1, kernel='rbf')
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
# accuracy
print("accuracy", metrics.accuracy_score(y_test, y_pred))

# precision
print("precision", metrics.precision_score(y_test, y_pred))

# recall/sensitivity
print("recall", metrics.recall_score(y_test, y_pred))
```
